## combine all col rows into single string of list

```python
combined_feature = df1["genres"]+" "+df1["keywords"]+" " +df1["tagline"]+" "+df1["cast"]+" "+df1["director"]
```

```python
combined_feature[0]
```

### Use count vectorize

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorize = CountVectorizer()
```

```python
feature_vec = vectorize.fit_transform(combined_feature)
```

```python
feature_vec.toarray()
```

```python
feature_vec.shape
```

```python
vectorize.get_feature_names_out()# give us all the words in col / unique values
```

## Cosine similarity return with col is import to others col

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(feature_vec)
similarity
```

```python
len(similarity)
```

```python
final_match = list(enumerate(similarity[index_of_title]))
final_match
```

# sorted( ) python build in function 

```python
from operator import itemgetter
```

```python
sorted_match = sorted(final_match , key=itemgetter(1), reverse=True)
```

```python
sorted_match[:5]
```

```python
print("Movie suggested for you: \n")
for i in range(1,6):
    index = sorted_match[i][0]
#     print(index)
    print(df.iloc[index]["title"])
```
